## Merge

Nesse notebook é feita a junção dos dados construídos com a malha geográfica

```python
from utils.load_csv import load_csv
from utils.load_shp import load_shp
from utils.save_shp import save_shp
```

```python
dados = load_csv("dados.csv")

distritos = load_shp("distritos.shp")
```

```python
mapa = distritos.merge(
    dados[["chave", "pontuacao"]],
    on="chave",
    how="left"
)

faltantes = mapa[mapa["pontuacao"].isna()]
if len(faltantes):
    print("Atenção, sem correspondência:", faltantes["nm_distrit"].tolist())
```

```python
save_shp(mapa, "mapa.shp")
```
