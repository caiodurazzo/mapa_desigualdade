## Tratamento

Nesse notebook os dados construídos e a malha geográfica são tratados

```python
import unicodedata
from utils.load_csv import load_csv
from utils.load_shp import load_shp
from utils.save_csv import save_csv
from utils.save_shp import save_shp
```

```python
dados = load_csv("dados.csv")

distritos = load_shp("distritos.shp")
```

```python
def normalizar(texto: str) -> str:
    texto = unicodedata.normalize("NFKD", str(texto))
    texto = texto.encode("ascii", "ignore").decode("utf-8")
    return texto.strip().upper()
```

```python
dados["chave"] = dados["distrito"].apply(normalizar)

distritos["chave"] = distritos["nm_distrit"].apply(normalizar)
```

```python
save_csv(dados, "dados.csv")

save_shp(distritos, "distritos.shp")
```
